```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
file_path="D:/221501124/OzoneHole_Data.csv"
df=pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['Hole Area'],y=df['Minimum Ozone'],hue=df['Year'])
plt.xlabel("Ozone Hole Area")
plt.ylabel("Minimum Ozone")
plt.title("Ozone Hole Area vs Minimum Ozone")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Hole Area'],y=df['Minimum Ozone'],palette='Blues')
plt.ylabel("Ozone hole Area ")
plt.xlabel("Year")
plt.title("Annual Ozone Hole Area")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Minimum Ozone'],bins=10,color='green')
plt.xlabel("Minimum Ozone Level")
plt.ylabel("Frequency")
plt.title("Distribution of Minimum Ozone Levels")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y=df["Hole Area"], color="orange")
plt.ylabel("Ozone Hole Area")
plt.title("Ozone Hole Area Distribution")
plt.grid(True)
plt.show()
```

```python
df["Rolling_Hole_Area"] = df["Hole Area"].rolling(window=5, min_periods=1).mean()

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["Year"], y=df["Rolling_Hole_Area"], color="purple")
plt.xlabel("Year")
plt.ylabel("5-Year Rolling Average")
plt.title("5-Year Rolling Average of Ozone Hole Area")
plt.grid(True)
plt.show()
```
