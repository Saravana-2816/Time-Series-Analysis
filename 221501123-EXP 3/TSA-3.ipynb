```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/221501124/EXP-3/OzoneHole_Data.csv")
```

```python
X = data['Year'].values.reshape(-1, 1)  # Feature: Year
y_area = data['Hole Area'].values  # Target 1: Hole Area
y_ozone = data['Minimum Ozone'].values  # Target 2: Minimum Ozone
```

```python
model_area = LinearRegression()
model_ozone = LinearRegression()

# Fit the models
model_area.fit(X, y_area)
model_ozone.fit(X, y_ozone)
```

```python
y_area_pred = model_area.predict(X)
y_ozone_pred = model_ozone.predict(X)

# Evaluate the models
mse_area = mean_squared_error(y_area, y_area_pred)
r2_area = r2_score(y_area, y_area_pred)

mse_ozone = mean_squared_error(y_ozone, y_ozone_pred)
r2_ozone = r2_score(y_ozone, y_ozone_pred)

print("Ozone Hole Area Prediction:")
print(f"MSE: {mse_area:.2f}, R²: {r2_area:.2f}")

print("\nMinimum Ozone Prediction:")
print(f"MSE: {mse_ozone:.2f}, R²: {r2_ozone:.2f}")
```

```python
plt.figure(figsize=(14, 6))

# Plot for Ozone Hole Area
plt.subplot(1, 2, 1)
plt.scatter(X, y_area, color='blue', label='Actual Area')
plt.plot(X, y_area_pred, color='red', label='Predicted Area')
plt.title('Ozone Hole Area Over Time')
plt.xlabel('Year')
plt.ylabel('Hole Area')
plt.legend()

# Plot for Minimum Ozone
plt.subplot(1, 2, 2)
plt.scatter(X, y_ozone, color='green', label='Actual Ozone')
plt.plot(X, y_ozone_pred, color='orange', label='Predicted Ozone')
plt.title('Minimum Ozone Over Time')
plt.xlabel('Year')
plt.ylabel('Minimum Ozone')
plt.legend()

plt.tight_layout()
plt.show()
```
